```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset into a DataFrame
data = pd.read_csv("fr-esr-patrimoine-immobilier-des-operateurs-de-l-enseignement-superieur (2).csv", delimiter=';', na_values='')

# Count the number of buildings by type of usage
building_usage_counts = data['type_bat'].value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
building_usage_counts.plot(kind='bar')
plt.xlabel('Type of Usage')
plt.ylabel('Number of Buildings')
plt.title('Number of Buildings by Type of Usage')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset into a DataFrame
data = pd.read_csv("fr-esr-patrimoine-immobilier-des-operateurs-de-l-enseignement-superieur.csv", delimiter=';', na_values='')

# Count the number of buildings by owner
building_owner_counts = data['proprietaire_bat'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
building_owner_counts.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Distribution of Building Owners')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset into a DataFrame
data = pd.read_csv("fr-esr-patrimoine-immobilier-des-operateurs-de-l-enseignement-superieur.csv", delimiter=';', na_values='')

# Create a box plot of energy consumption by building type
plt.figure(figsize=(10, 6))
data.boxplot(column='energie_valeur', by='type_bat', vert=False)
plt.xlabel('Energy Consumption (energie_valeur)')
plt.ylabel('Building Type')
plt.title('Distribution of Energy Consumption by Building Type')
plt.suptitle('')  # Remove the default title
plt.show()


# Energy Consumption vs. Surface Area
plt.scatter(df['SDP'], df['energie_valeur'])
plt.xlabel('SDP')
plt.ylabel('Energy Consumption')
plt.title('Energy Consumption vs. Surface Area')
plt.show()
```
